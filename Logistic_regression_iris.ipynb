```python
import pandas as pd
file_path = '/content/drive/My Drive/Data/iris/iris.data'

# Read the CSV file
df = pd.read_csv(file_path,names=["sepal_len","sepal_width","petal_len","petal_width","type"])
#df.reset_index(inplace=True)
iris = 'output.csv'
df.to_csv(iris, index=True)
# Display the first few rows of the DataFrame
df.head()
```

```python
df= pd.get_dummies(df, columns=['type'])
bool_cols = df.select_dtypes(include=['bool']).columns
df[bool_cols] = df[bool_cols].astype(int)
df.sample(5)
```

```python
x=df.iloc[:,0:4]
y=df.iloc[:,4:]
```

```python
import numpy as np
import random
x=np.array(x)
x= np.concatenate((np.ones((len(x), 1)), x),axis=1)
y=np.array(y)
m=np.random.permutation(len(x))
x=x[m]
y=y[m]
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x,  y, test_size=0.2, random_state=42)

from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

scaler.fit(x_train)
x_train= scaler.fit_transform(x_train)
x_test= scaler.fit_transform(x_test)
```

```python
class Classiifcation_FBGD:
  def __init__(self,epochs,lr):
    self.epochs=epochs
    self.lr=lr
    self.w=None
  def Loss(self,y,y1):
    return -np.sum(y*np.log(y1))

  def softmax(self,z):
    exp_z = np.exp(z - np.max(z, axis=1, keepdims=True))  # Stability trick
    return exp_z / np.sum(exp_z, axis=1, keepdims=True)

  def fit(self,x,y):
    L=[]
    interval=self.epochs//10
    self.w=np.random.randn(x.shape[1],y.shape[1])*(1/np.sqrt(x.shape[1]))
    for epoch in range(self.epochs):

      z=np.dot(x,self.w)
      y_pred=self.softmax(z)
      residual=y_pred-y
      gradient=np.matmul(x.T,residual)/len(y)
      self.w-=self.lr*gradient
      l=self.Loss(y,y_pred)
      L.append(l)
      if epoch % interval== 0:
        print(f"Epoch {epoch}, Loss: {l}")

    return L,self.w
  def predict(self,x):
    z=np.dot(x,self.w)
    return self.softmax(z)
```

```python
epochs=5000
lr=0.3
model=Classiifcation_FBGD(epochs,lr)
l,w=model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
y_pred = (y_pred > 0.5).astype(int)
#print(y_pred)
c=0
for i in range(len(y_test)):
  if np.array_equal(y_test[i], y_pred[i])==True:
    c+=1
print(c/len(y_test)*100)
```
